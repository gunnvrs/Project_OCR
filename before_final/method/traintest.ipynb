# ทำการเทรน
- หลังทำการรัน ย้ายไฟล์ gunn_trained_1.lstm ไปที่ไดเร็กทอรี Tesseract ที่ใช้เก็บไฟล์ .traineddata 
```
cp gunn_trained_1.lstm /Users/gunnviryasiri/pythainlp-data/tessdata/gunn.traineddata
```

````python
import os
import subprocess

def convert_bmp_to_png(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".bmp"):
            bmp_path = os.path.join(directory, filename)
            png_path = os.path.splitext(bmp_path)[0] + ".png"
            subprocess.run(["convert", bmp_path, png_path])

def generate_box_files(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".png"):
            png_path = os.path.join(directory, filename)
            base_name = os.path.splitext(png_path)[0]
            subprocess.run(["tesseract", png_path, base_name, "-l", "tha", "--psm", "6", "batch.nochop", "makebox"])

def train_model(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".png"):
            png_path = os.path.join(directory, filename)
            box_path = os.path.splitext(png_path)[0] + ".box"
            subprocess.run(["tesseract", png_path, box_path, "lstm.train"])

def extract_unicharset(directory):
    box_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith(".box")]
    subprocess.run(["unicharset_extractor", "--norm_mode", "3"] + box_files)

def combine_tessdata(tessdata_path, output_path):
    subprocess.run(["combine_tessdata", "-e", tessdata_path, output_path])

def main():
    # กำหนดเส้นทางโฟลเดอร์
    directory = "/Users/gunnviryasiri/Desktop/ocr/ThaiOCR/ThaiOCR-TrainigSet"  # แก้ไขเส้นทางนี้ให้ตรงกับตำแหน่งที่เก็บข้อมูล
    tessdata_path = "/Users/gunnviryasiri/pythainlp-data/tessdata/tha.traineddata"  # เส้นทางไปยังไฟล์ tha.traineddata
    # output_path = "tha.lstm"
    output_path = "gunn_trained_1.lstm"

    # แปลงไฟล์ BMP เป็น PNG
    convert_bmp_to_png(directory)

    # สร้างไฟล์ .box
    generate_box_files(directory)

    # ฝึกโมเดล
    train_model(directory)

    # สร้างไฟล์ unicharset
    extract_unicharset(directory)

    # สร้างไฟล์ training data
    combine_tessdata(tessdata_path, output_path)

    print("Training completed successfully!")

if __name__ == "__main__":
    main()
````

````python
import pytesseract
from PIL import Image
from datetime import datetime
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# ตั้งค่าเส้นทางไปยัง Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'/opt/homebrew/bin/tesseract'

# Path to the image file
image_path = '/Users/gunnviryasiri/Desktop/ocr/testfile/file1.jpg'

# สร้างชื่อไฟล์จากวันที่และเวลา
current_time = datetime.now().strftime("%Y%m%d_%H%M%S")
output_text_file = f'/Users/gunnviryasiri/Desktop/ocr/results/ocr_output_{current_time}.txt'

# Load the image
image = Image.open(image_path)

# Convert the image to grayscale
gray = image.convert('L')

# Perform OCR on the image using Thai and English languages
custom_config = r'--tessdata-dir "/Users/gunnviryasiri/pythainlp-data/tessdata" --oem 3 --psm 6'
text = pytesseract.image_to_string(gray, config=custom_config, lang='tha+eng')

# ลบช่องว่างทั้งหมดออกจากข้อความที่ได้
text_no_spaces = text.replace(" ", "")

# Save the cleaned text to a file
with open(output_text_file, 'w', encoding='utf-8') as f:
    f.write(text_no_spaces)

print(f'Text extracted, cleaned, and saved to {output_text_file}')

# อ่านข้อความจากไฟล์ FORTEST.txt
with open('/Users/gunnviryasiri/Desktop/ocr/results/FORTEST.txt', 'r', encoding='utf-8') as f:
    for_test_text = f.read().replace(" ", "")

# อ่านข้อความที่ได้จาก OCR
with open(output_text_file, 'r', encoding='utf-8') as f:
    ocr_text = f.read()

# สร้าง list ของตัวอักษรจากข้อความ
for_test_chars = list(for_test_text)
ocr_chars = list(ocr_text)

# ปรับขนาดของลิสต์ให้ตรงกัน
max_len = max(len(for_test_chars), len(ocr_chars))
for_test_chars = for_test_chars + [''] * (max_len - len(for_test_chars))
ocr_chars = ocr_chars + [''] * (max_len - len(ocr_chars))

# สร้าง confusion matrix
cm = confusion_matrix(for_test_chars, ocr_chars, labels=np.unique(for_test_chars + ocr_chars))

# แสดง confusion matrix
print("Confusion Matrix:")
print(cm)

# แสดงผล classification report
print("Classification Report:")
print(classification_report(for_test_chars, ocr_chars))
````


## ความหมาย
- Precision: ความถูกต้องของการทำนายว่าเป็นตัวอักษรที่เฉพาะเจาะจง
- Recall: ความสามารถในการจับคู่ตัวอักษรที่เป็นจริงในข้อความ
- F1-score: ความสมดุลระหว่าง Precision และ Recall
- Support: จำนวนตัวอย่างที่มีตัวอักษรนั้น

- Precision และ Recall: ตัวอักษรบางตัวมี precision และ recall ต่ำ อาจหมายความว่าการ OCR ยังไม่สามารถจับคู่ได้ดี หรือมีตัวอักษรที่ไม่ค่อยเห็นในข้อมูลฝึกหัด

````python
import pytesseract
from PIL import Image
import cv2
from datetime import datetime

# ตั้งค่าเส้นทางไปยัง Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'/opt/homebrew/bin/tesseract'

# Path to the image file
image_path = '/Users/gunnviryasiri/Desktop/ocr/testfile/file1.jpg'

# Path to the reference file
reference_text_file = '/Users/gunnviryasiri/Desktop/ocr/results/FORTEST.txt'

# สร้างชื่อไฟล์จากวันที่และเวลา
current_time = datetime.now().strftime("%Y%m%d_%H%M%S")
output_text_file = f'/Users/gunnviryasiri/Desktop/ocr/results/ocr_output_{current_time}.txt'

# Load the image
image = Image.open(image_path)

# Convert the image to grayscale
gray = image.convert('L')

# Perform OCR on the image using Thai and English languages
custom_config = r'--tessdata-dir "/Users/gunnviryasiri/pythainlp-data/tessdata" --oem 3 --psm 6'
text = pytesseract.image_to_string(gray, config=custom_config, lang='tha+eng')

# ลบช่องว่างทั้งหมดออกจากข้อความที่ได้
text_no_spaces = text.replace(" ", "")

# Save the cleaned text to a file
with open(output_text_file, 'w', encoding='utf-8') as f:
    f.write(text_no_spaces)

# อ่านข้อความที่ถูกต้องจากไฟล์ reference
with open(reference_text_file, 'r', encoding='utf-8') as f:
    correct_text = f.read().replace(" ", "")

# เปรียบเทียบข้อความและสร้าง confusion matrix
tp = tn = fp = fn = 0

# ตรวจสอบความยาวของข้อความทั้งสองก่อนเปรียบเทียบ
min_length = min(len(text_no_spaces), len(correct_text))

for i in range(min_length):
    if text_no_spaces[i] == correct_text[i]:
        tp += 1  # True Positive
    else:
        fp += 1  # False Positive

# กรณีข้อความที่ OCR สั้นกว่าข้อความที่ถูกต้อง
if len(text_no_spaces) < len(correct_text):
    fn += len(correct_text) - len(text_no_spaces)  # False Negative: ข้อความหายไป

# กรณีข้อความที่ OCR ยาวกว่าข้อความที่ถูกต้อง
if len(text_no_spaces) > len(correct_text):
    fp += len(text_no_spaces) - len(correct_text)  # False Positive: ข้อความเกินมา

# แสดงผลลัพธ์ Confusion Matrix
print(f"True Positive (TP): {tp}")
print(f"False Positive (FP): {fp}")
print(f"False Negative (FN): {fn}")
# เราไม่สามารถตรวจสอบ True Negative ได้ (เพราะไม่มีข้อมูลที่ไม่ได้มีอยู่จริงในภาพ)

accuracy = tp / (tp + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)

print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')


print(f'Text extracted, cleaned, and saved to {output_text_file}')
````
